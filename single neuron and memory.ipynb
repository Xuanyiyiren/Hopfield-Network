```python
import numpy as np
import matplotlib.pyplot as plt

import scipy as sp
```

```python

activation=lambda x:np.tanh(x)

def input(t):
    if (t>1 and t<2):
        return 2
    elif (t>11 and t<12):
        return -2
    else:
        return 0
    # return 0

def recurrent(x,t,lam):
    dxdt=-x+activation(lam*x+input(t))
    return dxdt

t=np.linspace(0,25,1000)
y0=0

lam=[0.1,1.9]
# lam=[0.1,1.9]
for i in lam:
    sol = sp.integrate.odeint(recurrent,y0,t,args=((i,)))
    plt.plot(t, sol[:, 0], label=f'$\\lambda={i}$')
plt.plot(t,0.3*np.frompyfunc(input,1,1)(t),label='Input:$h(t)$')

plt.legend(loc='best')
plt.xlabel('t/$\\tau_n$')
plt.title('$\\nu(t)$')
# plt.yscale('log')
plt.grid()
plt.savefig('../fig/htandfixedpoint.pdf')
plt.show()
```
